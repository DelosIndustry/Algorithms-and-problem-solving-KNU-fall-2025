```python
def insertion_sort(arr):
    # 두 번째 요소부터 시작!
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        # key보다 큰 요소들을 한 칸씩 뒤로 이동
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        print(f"Step {i}: {arr}")
    return arr
```

```python
arr = [64, 34, 25, 12, 22, 11, 90]
insertion_sort(arr)
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        print(f"Step {i + 1}: {arr}")
    return arr
```

```python
arr = [64, 34, 25, 12, 22, 11, 90]
selection_sort(arr)
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
        print(f"Step {i + 1}: {arr}")
    return arr
```

```python
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(arr)
```

```python
def quick_sort(arr, low, high):
    if low < high:
        # Partition 함수를 통해 피벗의 위치(p)를 확정
        p = partition(arr, low, high)
        
        # 피벗을 제외한 왼쪽 부분 정렬
        quick_sort(arr, low, p - 1)
        # 피벗을 제외한 오른쪽 부분 정렬
        quick_sort(arr, p + 1, high)

def partition(arr, low, high):
    pivot = arr[low] # 피벗을 가장 왼쪽 요소로 설정
    left = low + 1
    right = high
    
    while True:
        # left가 right보다 커지거나, pivot보다 큰 값을 만날 때까지 전진
        while left <= right and arr[left] <= pivot:
            left += 1
            
        # right가 left보다 작아지거나, pivot보다 작은 값을 만날 때까지 후진
        while left <= right and arr[right] >= pivot:
            right -= 1
            
        # 엇갈렸다면 루프 종료
        if left > right:
            break
            
        # 엇갈리지 않았다면 두 값을 교환 (Swap)
        arr[left], arr[right] = arr[right], arr[left]
    
    # 피벗을 자신의 자리(right 위치)로 이동
    arr[low], arr[right] = arr[right], arr[low]
    
    return right # 피벗의 최종 인덱스 반환
```

```python
arr = [64, 34, 25, 12, 22, 11, 90]
quick_sort(arr, 0, len(arr) - 1)
print(f"Sorted array: {arr}")
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
            
    result.extend(left[i:])
    result.extend(right[j:])
    return result 
```

```python
arr = [64, 34, 25, 12, 22, 11, 90]
merged_arr = merge_sort(arr)
print(f"Sorted array: {merged_arr}")
```

```python
import heapq

def heap_sort(arr):
    heap = []
    for element in arr:
        heapq.heappush(heap, element)
        
    sorted_arr = []
    while heap:
        sorted_arr.append(heapq.heappop(heap))
    return sorted_arr
```

```python
arr = [64, 34, 25, 12, 22, 11, 90]
heap_sorted_arr = heap_sort(arr)
print(f"Sorted array: {heap_sorted_arr}")
```

```python
def topological_sort(graph):
    visited = set()
    stack = []
    
    def dfs(node):
        visited.add(node)
        
        if node in graph:
            for neighbor in graph[node]:
                if neighbor not in visited:
                    dfs(neighbor)
        stack.append(node)
        
    all_nodes = set(graph.keys())
    for values in graph.values():
        for v in values:
            if v not in all_nodes:
                all_nodes.add(v)
    for node in set(all_nodes):
        if node not in visited:
            dfs(node)
    
    return stack[::-1]  # 역순으로 반환
```

```python
dag_graph = {'Underwear': ['Pants', 'Shoes'],
        'Pants': ['Belt', 'Shoes'],
        'Socks': ['Shoes'],
        'Shirt': ['Belt', 'Tie'],
        'Belt': ['Jacket'],
        'Tie': ['Jacket'],
        'Shoes': [],
        'Jacket': []
    }
topological_order = topological_sort(dag_graph)
print(f"Topological Sort Order: {topological_order}")
```
